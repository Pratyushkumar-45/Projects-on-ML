```python
# separating the data for analysis
legit=bankdata[bankdata.Class==0]
fraud=bankdata[bankdata.Class==1]
```

To make the data balanced we have used undersampling technique

building a sample dataset from orignal dataset containing similar distribution of fraud and legit transaction

```python
legit_sample=legit.sample(n=492)
```

concatening two data frames

```python
newdata=pd.concat([legit_sample,fraud],axis=0)
```

```python
newdata.head(10)
```

```python
 newdata['Class'].value_counts()
```

 Now the data is balanced

## spliting the data

Now dividing the data for logistic regression algorithm into feature and target

```python
X=newdata.drop(columns='Class',axis=1)
Y=newdata['Class']
```

```python
print(X)
```

```python
print(Y)
```

```python
X_train, X_test ,Y_train, Y_test =train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

## model 

## logistic regression

```python
model=LogisticRegression()
```

```python
model.fit(X_train, Y_train)
```

## model evaluation

```python
X_train_prediction=model.predict(X_train)
training_data_accuracy=accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on Training Data:',training_data_accuracy)
```

```python
X_test_prediction=model.predict(X_test)
test_data_accuracy=accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy scre on test data:',test_data_accuracy)
```

Thus It can be seen that the accuracy score of both test and train data is comparable and good
